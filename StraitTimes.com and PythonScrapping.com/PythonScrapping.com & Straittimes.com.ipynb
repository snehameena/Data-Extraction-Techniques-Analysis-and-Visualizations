```python
_news_url = "https://www.straitstimes.com/"
```

```python
html = uReq(_news_url)
page_soup = soup(html.read(), 'html.parser');
print(page_soup.h1)
```

```python
a_container = page_soup.findAll('div', attrs={'class':'view-content'})
print(a_container)
```

```python
_news_list = []
a_container1 = page_soup.findAll('span',attrs={'class':'story-headline'})
for a in a_container1:
    _news_list.append((a.find('a').text))
```

```python
_news_list
```

```python
_main_word_list = []
for line in _news_list:
    token_list1 = word_tokenize(line)
    for token1 in token_list1:
        _main_word_list.append(token1)

_main_word_list
```

```python
from nltk.book import *
```

```python
_preprocessed_text = []
for _w2 in _main_word_list:
    if(_w2.isalnum()):
        _preprocessed_text.append(_w2)
        
t2 = nltk.Text(_preprocessed_text)
t2
```
